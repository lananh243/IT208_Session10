```python
# Bài 1:
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

input_data = np.random.normal(7, 1.5, 200)

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 6))

sns.histplot(
    input_data,
    bins=20,
    kde=True,
    color="skyblue",
)

plt.title("Phân phối điểm thi (Histogram + KDE)")
plt.xlabel("Điểm số")
plt.ylabel("Tần suất")

plt.show()
```

```python
# Bài 2:
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "subject": ['Toán', "Lý", "Hóa"],
    "score": [9, 7.6, 8]
}

df = pd.DataFrame(data)

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 6))

sns.displot(df, x='score', hue='subject', kind='kde')

plt.show()
```
