```python
# Bài 1:
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

input_data = np.random.normal(7, 1.5, 200)

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 6))

sns.histplot(
    input_data,
    bins=20,
    kde=True,
    color="skyblue",
)

plt.title("Phân phối điểm thi (Histogram + KDE)")
plt.xlabel("Điểm số")
plt.ylabel("Tần suất")

plt.show()
```

```python
# Bài 2:
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

subject = ['Toán', "Lý", "Hóa"],
score = np.random.normal(7, 1.5, 300)
subject_col = np.repeat(subject, 100)  

df = pd.DataFrame({
    "subject": subject_col,
    "score": score
})

sns.set_theme(style="whitegrid")
sns.displot(df, x='score', hue='subject', kind='kde')

plt.show()
```

```python
# Bài 3:
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

hours = np.arange(1, 11)                         
noise = np.random.normal(0, 1, len(hours))        
score = hours * 1.2 + noise   

df = pd.DataFrame({
    "hours": hours,
    "score": score
})

sns.set_theme(style="whitegrid")

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x="hours", y="score")

plt.title("Mối quan hệ giữa số giờ học và điểm số")
plt.xlabel("Số giờ học")
plt.ylabel("Điểm số")

plt.show()
```

```python
# Bài 4:
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

classes = ["A", "B", "C"]
class_col = np.repeat(classes, 50)
scores = np.concatenate([
    np.random.normal(8, 1, 50),
    np.random.normal(7, 1, 50),
    np.random.normal(7.5, 1, 50)
])

df = pd.DataFrame({
    "class": class_col,
    "score": scores
})

sns.set_theme(style="whitegrid")

sns.barplot(df, x="class", y="score", errorbar="sd")

plt.title("So sánh điểm trung bình giữa các lớp")
plt.xlabel("Lớp")
plt.ylabel("Điểm trung bình")

plt.show()
```

```python
# Bài 5: 
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

major = ["CNTT"] * 50 + ["Kinh tế"] * 50 + ["Ngôn ngữ"] * 50
scores = np.concatenate([
    np.random.normal(8, 1, 50),
    np.random.normal(9, 1, 50),
    np.random.normal(6, 1, 50)
])

df = pd.DataFrame({
    "major": major,
    "score": scores
})
sns.set_theme(style="whitegrid")

# BOX PLOT
# plt.figure(figsize=(8, 6))
# sns.boxplot(data=df, x="major", y="score")
# plt.title("Boxplot điểm theo ngành")
# plt.show()

# VIOLIN PLOT
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x="major", y="score")
plt.title("Violinplot điểm theo ngành")
plt.show()

# Nhóm có điểm phân tán rộng: Mức phân tán tương đương nhau, không nhóm nào rộng vượt trội.
# Nhóm có điểm tập trung hơn: Vì độ phân tán bằng nhau nên mức độ tập trung tương đương.
# Nhóm có trung vị (median) cao hơn:
# Nhóm Kinh Tế có trung vị cao nhất (≈ 9).
# Nhóm CNTT có trung vị đứng thứ hai (≈ 8).
# Nhóm Ngôn Ngữ có trung vị thấp nhất (≈ 6).
```

```python
# Bài 6:
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

day = np.arange(1, 31)                         
noise = np.random.normal(0, 1, len(day))        
sales = day + noise

df = pd.DataFrame({
    "day": day,
    "sale": sales
})

sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 6))

sns.lineplot(data=df, x="day", y="sale", marker="o")
plt.title("Doanh thu từng ngày (Lineplot)")
plt.show()

sns.regplot(data=df, x="day", y="sale")
plt.title("Xu hướng doanh thu theo ngày (Regplot)")
plt.show()

# Doanh thu co xu huong ngay cang tang
# Doanh thu có dao động nhẹ, không hoàn toàn đều do có noise
# Đường hồi quy đi lên, chứng tỏ doanh thu tổng thể tăng dần theo ngày
```

```python
# Bài 7:
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

hours = np.random.uniform(1, 10, 500)
noise = np.random.normal(0, 2, 500)  
score = hours * 1.5 + noise

major = np.random.choice(["A", "B", "C"], 500)

df = pd.DataFrame({
    "hours": hours,
    "score": score,
    "major": major
})

sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 5))

# Histplot
sns.histplot(df["score"], bins=30, kde=True)
plt.title("Phân phối điểm")
plt.show()

# Scatterplot
sns.scatterplot(data=df, x="hours", y="score")
plt.title("Mối quan hệ giữa giờ học và điểm")
plt.show()

# Boxplot theo ngành
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x="major", y="score")
plt.title("So sánh điểm giữa các ngành")
plt.show()
```

```python
# Bài 8:

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd


months = np.arange(1, 13)
categories = ["Shoes", "Clothes", "Bags"]

data = []
for cat in categories:
    revenue = np.random.randint(100, 300, size=12)
    for month, rev in zip(months, revenue):
        data.append({"month": month, "revenue": rev, "category": cat})

df = pd.DataFrame(data)

sns.set_theme(style="whitegrid")
plt.figure(figsize=(8, 5))

sns.lineplot(data=df, x="month", y="revenue", hue="category")
plt.title("Doanh thu theo tháng bởi Category")
plt.xticks(months)
plt.show()

sns.violinplot(data=df, x="category", y="revenue")
plt.title("Phân phối doanh thu theo danh mục")
plt.show()

sns.regplot(data=df, x="month", y="revenue", scatter_kws={"s":50}, line_kws={"color":"red"})
plt.title("Xu hướng doanh thu chung")
plt.show()

# Nhóm Shoes doanh thu ổn định nhất.
# Nhóm Bags biến động mạnh nhất.
# Nhóm Clothes ở mức trung bình, biến động vừa phải.
# Xu hướng chung không rõ rệt, doanh thu dao động quanh mức trung bình.
# Có khác biệt về biến động giữa ba nhóm, Shoes thường có doanh thu cao hơn.
```
